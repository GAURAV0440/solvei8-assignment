```python
import pandas as pd
df = pd.read_csv("hotel_bookings.csv")
# Show the first 10 rows
df.head(10)
```

```python
# Basic info about dataset
print("Dataset Info:")
df.info()

# Checking missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Checking basic stats
print("\nSummary Statistics:")
df.describe()
```

```python
# Fill missing values
df.fillna({
    'children': 0,
    'country': 'Unknown',
    'agent': 0,
    'company': 0
}, inplace=True)

# Verify again
print(df.isnull().sum())
```

```python
# Total nights stayed = weekday + weekend nights
df['total_nights'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights']

# Total guests = adults + children + babies
df['total_guests'] = df['adults'] + df['children'] + df['babies']

# Convert reservation date to datetime format
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])

# Preview the new columns
df[['total_nights', 'total_guests', 'reservation_status_date']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate revenue (only from successful bookings)
df['revenue'] = df['adr'] * df['total_nights']
df['revenue'] = df['revenue'].where(df['is_canceled'] == 0, 0)

# Revenue trend over months
monthly_revenue = df[df['is_canceled'] == 0].groupby(['arrival_date_year', 'arrival_date_month'])['revenue'].sum().reset_index()

# Sort months for proper ordering
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
monthly_revenue['arrival_date_month'] = pd.Categorical(monthly_revenue['arrival_date_month'], categories=month_order, ordered=True)
monthly_revenue = monthly_revenue.sort_values(['arrival_date_year', 'arrival_date_month'])

# Plot revenue trend
plt.figure(figsize=(12, 5))
sns.lineplot(data=monthly_revenue, x='arrival_date_month', y='revenue', hue='arrival_date_year', marker='o')
plt.title('Monthly Revenue Trend')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Cancellation rate
cancel_rate = df['is_canceled'].mean() * 100
print(f"Cancellation Rate: {cancel_rate:.2f}%")
```

```python
# Top 10 countries by number of bookings
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.title('Top 10 Countries by Number of Bookings')
plt.xlabel('Country Code')
plt.ylabel('Number of Bookings')
plt.show()

# Lead time distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['lead_time'], bins=50, kde=True)
plt.title('Booking Lead Time Distribution')
plt.xlabel('Days between Booking & Arrival')
plt.ylabel('Frequency')
plt.show()
```

```python
# Repeated vs New Guests
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='is_repeated_guest')
plt.title('Repeated vs New Guests')
plt.xticks([0, 1], ['New', 'Repeated'])
plt.show()

# Special Requests Distribution
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='total_of_special_requests')
plt.title('Special Requests Distribution')
plt.xlabel('Number of Requests')
plt.ylabel('Booking Count')
plt.show()

# Car Parking Usage
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='required_car_parking_spaces')
plt.title('Car Parking Space Requirement')
plt.xlabel('Spaces Required')
plt.ylabel('Number of Bookings')
plt.show()

# Reserved vs Assigned Room Type match
match = (df['reserved_room_type'] == df['assigned_room_type']).mean() * 100
print(f"Percentage of guests who got the room type they reserved: {match:.2f}%")
```

```python
# Step 1: Imports
import pandas as pd
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

# Step 2: Load your dataset (make sure CSV is in your working directory)
df = pd.read_csv("hotel_bookings.csv")

# Step 3: Recreate the helper columns
df['total_nights'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights']
df['total_guests'] = df['adults'] + df['children'].fillna(0) + df['babies']

# Step 4: Prepare text for embedding
rag_df = df[['hotel', 'arrival_date_year', 'arrival_date_month', 'country', 'lead_time',
             'is_canceled', 'adr', 'total_nights', 'total_guests']].copy()

rag_df['text'] = rag_df.apply(
    lambda row: f"{row['hotel']} hotel booking from {row['country']} for {int(row['total_guests'])} guest(s) staying {int(row['total_nights'])} night(s) in {row['arrival_date_month']} {row['arrival_date_year']}. Lead time: {row['lead_time']} days. Price: {row['adr']}. Canceled: {bool(row['is_canceled'])}.",
    axis=1
)

# Step 5: Sample 2000 rows for speed
sample_rag = rag_df.sample(2000, random_state=42).reset_index(drop=True)
texts = sample_rag['text'].tolist()

# Step 6: Load model and generate embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(texts, convert_to_numpy=True, normalize_embeddings=True, batch_size=64)
embeddings = embeddings.astype('float32')

# Step 7: Create FAISS index
index = faiss.IndexFlatL2(embeddings.shape[1])
index.add(embeddings)

print(f"✅ FAISS index created with {index.ntotal} documents.")
```

```python
# Function to search similar records using FAISS
def search_similar(query, top_k=3):
    # Embed the query
    query_embedding = model.encode([query], convert_to_numpy=True, normalize_embeddings=True).astype('float32')
    
    # Search FAISS index
    distances, indices = index.search(query_embedding, top_k)
    
    # Show top matching texts
    results = [sample_rag.iloc[i]['text'] for i in indices[0]]
    return results

# Example question
query = "Show me bookings from Portugal with more than 2 guests"

# Search and print answers
matches = search_similar(query)
for i, result in enumerate(matches, 1):
    print(f" Match {i}:\n{result}\n")
```
